```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numberOfEnergyLevels = 1000
eIdxs = np.arange(numberOfEnergyLevels)

temperature = 20.0
beta = 1.0 / temperature

distribution = np.zeros(numberOfEnergyLevels)
```

```python
distribution = np.exp(-beta * eIdxs)
def linear(beta):
    dist = np.exp(-beta * eIdxs)
    return dist / dist.sum()

def degeneracy(beta):
    dist = (eIdxs + 1) * np.exp(-beta * eIdxs)
    return dist / dist.sum()

def inertia(beta):
    dist = (2 * eIdxs + 1) * np.exp(-beta * (0.5 * eIdxs * (eIdxs + 1)))
    return dist / dist.sum()

#distribution = (eIdxs + 1) * np.exp(-beta * eIdxs)
#distribution = (2 * eIdxs + 1) * np.exp(-beta * (0.5 * eIdxs * (eIdxs + 1)))
```

```python
distribution /= distribution.sum()
```

```python
fig, ax = plt.subplots()
for beta in [2.0 ,1.0, 0.5, 0.2]:
    ax.plot(eIdxs, linear(beta), label=f"T = {1/beta}")
ax.set_xlim(0, 10)
ax.legend()
```

```python
fig, ax = plt.subplots()
for beta in [2.0 ,1.0, 0.5, 0.2]:
    ax.plot(eIdxs, degeneracy(beta), label=f"T = {1/beta}")
ax.set_xlim(0, 10)
ax.legend()
```

```python
fig, ax = plt.subplots()
for beta in [2.0 ,1.0, 0.5, 0.2]:
    ax.plot(eIdxs, inertia(beta), label=f"T = {1/beta}")
ax.set_xlim(0, 10)
ax.legend()
```
